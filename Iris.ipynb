# **Importing Libraries**

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
path = kagglehub.dataset_download("jillanisofttech/iris-dataset-uci")

print("Path to dataset files:", path)

print(os.listdir(path))
iris_data = os.path.join(path,"Iris.csv")

iris = pd.read_csv(iris_data)
print(iris.head())
```

```python
print(iris.describe())
print(iris.shape)
```

```python
print(iris.isnull().sum())
```

```python
print(iris['Species'].value_counts())
```

```python
ax = sns.barplot(x = 'Species', y = 'SepalLengthCm', hue = 'Species', data = iris, ci= None)
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=2)
plt.title('Sepal Length by Species')
plt.show()
```

```python
sns.violinplot(x = 'Species', y = 'SepalWidthCm', data = iris)
plt.title('Sepal Width by Species')
plt.show()
```

```python
numeric = iris.select_dtypes(include = ['number'])
corr_mat = numeric.corr(method = 'pearson')

plt.figure(figsize = (10,8))
plt.title('Correlation Matrix')
sns.heatmap(corr_mat, annot = True , cmap = 'coolwarm')
plt.show()
```

```python
sns.kdeplot(x = 'SepalLengthCm', hue = 'Species', data = iris)
plt.title('Sepal Length by Species')
plt.show()
```

```python
sns.boxplot(x = 'Species', y = 'SepalWidthCm', data = iris)
plt.title('Sepal Width by Species')
plt.show()
```

```python
sns.swarmplot(x = 'Species', y = 'SepalWidthCm', data = iris)
plt.title('Sepal Width by Species')
plt.show()
```

```python
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', data = iris)
plt.title('Sepal Length vs Width by Species')
plt.show()
```

```python
sns.lineplot(x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', data = iris)
plt.title('Sepal Length vs Width by Species')
plt.show()
```

```python
sns.countplot(x = 'Species', data = iris)
plt.title('Count of Species')
plt.show()
```
